```python

from Extraction import Extraction 
from sqlalchemy import text 

table_name= 'order_items'
engine= 'mysql+pymysql://root:seif@localhost:3306/source'
dbtype= 'sql'
api_url= 'https://openexchangerates.org/api/latest.json?app_id=a92f8bad8e044bc79949a676886da2c8'

extract= Extraction(table_name , engine , dbtype)

eng=extract.create_engine()
typ=extract.dbtype_check()
df = extract.read_from_sql()

df["source"] = "mysql"
td=extract.get_today()
df["Extracted_date"] = td
dollar_value=extract.extract_value_from_api(api_url)

df["list_price"]= df["list_price"].astype(float) / dollar_value

table_name2= 'order_items'
engine2= 'mysql+pymysql://root:seif@localhost:3306/Extraction'
extract= Extraction(table_name2 , engine2 , dbtype)
extract.load_to_sql(df, table_name2, engine2)

```
