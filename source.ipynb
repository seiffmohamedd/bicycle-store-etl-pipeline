```python

from loading import loading 
from sqlalchemy import text 

csv_file_name= 'customers.csv'
engine= 'mysql+pymysql://root:seif@localhost:3306/source'

load= loading(csv_file_name , engine)

eng=load.create_engine()
df = load.read_from_csv()
table_name = load.trans_csv_name()
# tables = load.insp_table()
# print(tables)
with eng.connect() as conn:
    conn.execute(text("""
        CREATE Table customers (
            customer_id INT PRIMARY KEY, 
            first_name VARCHAR(50), 
            last_name VARCHAR(50), 
            phone varchar(50),
            email varchar(50),
            street varchar(50),
            city varchar(50),
            state varchar(5),
            zip_code varchar(10)
        )
    """))

load.save_to_csv(df)

```
